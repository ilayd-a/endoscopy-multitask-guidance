```python
import pandas  as pd
import numpy as np
from PIL import Image
```

```python
import os

# Define a folder path where your images are stored
folder_path = '`/dataset' #Update this to the actual path where your images are stored
image_list = []

# Read all images in the folder and convert to arrays
for image_file in os.listdir(folder_path):
    if image_file.endswith(('.png', '.jpg', '.jpeg')):
        img_path = os.path.join(folder_path, image_file)
        img = Image.open(img_path)
        # Convert image to a NumPy array
        img_array = np.array(img)
        image_list.append({'filename': image_file, 'image_data': img_array})
```

```python
# Create a DataFrame from the list of image data
df_hcg = pd.DataFrame(image_list)
print(df_hcg.shape)
```

```python
# Display the columns of the DataFrame
dimensions = df_hcg.shape
print(f"Dimensions (rows, columns): {dimensions}")
```

```python
print(df_hcg.columns)
```

```python
# Display the image data for the first few rows
print(df_hcg.image_data)
```

```python
# Drop the 'filename' column if it's not needed for further processing
df_hcg = df_hcg.drop(columns='filename')
```
