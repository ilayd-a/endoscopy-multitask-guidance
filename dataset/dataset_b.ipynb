```python
import pandas  as pd
import numpy as np
from PIL import Image
```

```python
import os
from pathlib import Path

# Define a folder path where your images are stored
BASE_DIR    = Path.cwd()
folder_path = Path("/Users/haala/endoscopy-multitask-guidance/dataset/HGC")

#BASE_DIR / "EndoscopicBladderTissue" / "HGC"

if not folder_path.exists():
    raise FileNotFoundError(
        f"❌ HGC folder not found at {folder_path}\n"
        "Download from Kaggle: https://www.kaggle.com/datasets/ingtborzo/endoscopic-bladder-tissue-classification\n"
        "Then place it at: dataset/EndoscopicBladderTissue/HGC/"
    )

image_list =[]
# Read all images in the folder and convert to arrays
for image_file in os.listdir(folder_path):
    if image_file.endswith(('.png', '.jpg', '.jpeg')):
        img_path = os.path.join(folder_path, image_file)
        img = Image.open(img_path)
        # Convert image to a NumPy array
        img_array = np.array(img)
        image_list.append({'filename': image_file, 'image_data': img_array})
```

```python
# Create a DataFrame from the list of image data
df_hcg = pd.DataFrame(image_list)
print(df_hcg.shape)
```

```python
# Display the columns of the DataFrame
dimensions = df_hcg.shape
print(f"Dimensions (rows, columns): {dimensions}")
```

```python
print(df_hcg.columns)
```

```python
# Display the image data for the first few rows
print(df_hcg.image_data)
```

```python
# Drop the 'filename' column if it's not needed for further processing
df_hcg = df_hcg.drop(columns='filename')
```

```python
# Repeat the same process for the second folder
folder_path = Path("/Users/haala/endoscopy-multitask-guidance/dataset/LGC")

#BASE_DIR / "EndoscopicBladderTissue" / "LGC"

if not folder_path.exists():
    raise FileNotFoundError(
        f"❌ LGC folder not found at {folder_path}\n"
        "Download from Kaggle: https://www.kaggle.com/datasets/ingtborzo/endoscopic-bladder-tissue-classification\n"
        "Then place it at: dataset/EndoscopicBladderTissue/LGC/"
    )

image_list = []

# Read all images in the folder and convert to arrays
for image_file in os.listdir(folder_path):
    if image_file.endswith(('.png', '.jpg', '.jpeg')):
        img_path = os.path.join(folder_path, image_file)
        img = Image.open(img_path)
        # Convert image to a NumPy array
        img_array = np.array(img)
        image_list.append({'filename': image_file, 'image_data': img_array})
```

```python
# Create a DataFrame from the list of image data
df_lcg = pd.DataFrame(image_list)
print(df_lcg.shape)
```

```python
df_lcg.columns
```

```python
df_lcg = df_lcg.drop(columns='filename')
```

```python
df = pd.concat([df_hcg, df_lcg], axis=0)
```

```python
from sklearn.model_selection import train_test_split

# First split: train vs temp
train_df, temp_df = train_test_split(
    df,
    test_size=0.30,
    random_state=42
)

# Second split: validation vs test
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    random_state=42
)

print("Train size:", train_df.shape)
print("Validation size:", val_df.shape)
print("Test size:", test_df.shape)
```
